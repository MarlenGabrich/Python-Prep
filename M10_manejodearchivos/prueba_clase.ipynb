### Manejo de archivos y web scrapping

Archivos secuenciales: Info almacenada en registros contiguos. Archivos de texto simple

Archivos Binarios: Codificados en lenguaje binario. Requieren procesamiento especifico para su lectura

```python
import os
```

```python
ingredientes = ['cebolla','tomate','quesito']
archivo = open('datitosh.txt','w')
```

```python
for elemento in ingredientes:
    archivo.write(elemento+'\n')
```

```python
archivo.close()
```

```python
archivo = open('datitosh.txt','r')
```

```python
for linea in archivo:
    print(linea)
```
